```python
#TASK3
dict1={}
string1=input("Please enter your dictionary key and values: ").split(',')
for i in string1:
  g = i.split(" : ")
  dict1[g[0]]=g[1]
d2={}
for k,v in dict1.items():
  if v not in d2.keys():
    d2[v]=[k]
  else:
    d2[v].append(k)
print(d2)
```
