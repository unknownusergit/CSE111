```python
# Task 2
class Assassin:
    assassin=0
    def __init__(self,name,failure):
        self.name=name
        self.failure=failure
        Assassin.assassin+=1
    def printDetails(self):
        print('Name:',self.name)
        print('Success rate:',self.failure)
        print('Total number of Assassin:',Assassin.assassin)
    @classmethod
    def failureRate(cls,name,fail):
        successrate=str(100-fail)+'%'
        cls.name=name
        return cls(name,successrate)
    @classmethod
    def failurePercentage(cls,name,fail):
        success_rate=str(100-fail)+'%'
        cls.name=name
        return cls(name,success_rate)
        
# Write your code here
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 5
class Employee:
    def __init__(self,name,workingPeriod):
        self.name=name
        self.workingPeriod=workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        name=name
        workingPeriod=2021-year
        return cls(name,workingPeriod)
    @staticmethod
    def experienceCheck(period,gender):
        if gender.lower()=='male':
            if period<3:
                return('He is not experienced')
            else:
                return('He is experienced')
        elif gender.lower()=='female':
            if period<3:
                return('She is not experienced')
            else:
                return('She is experienced')
        
# Write your code here
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 7
class Cat:
    Number_of_cats=0
    def __init__(self,color,activity):
        self.color=color
        self.activity=activity
        Cat.Number_of_cats+=1
    def printCat(self):
        print(self.color,'cat is',self.activity)
    @classmethod
    def no_parameter(cls):
        color='White'
        activity='sitting'
        return cls(color,activity)
    @classmethod
    def first_parameter(cls,color):
        clr=color
        activity='sitting'
        return cls(clr,activity)
    @classmethod
    def second_parameter(cls,activity):
        clr='White'
        act=activity
        return cls(clr,act)
    def changeColor(self,color):
        self.color=color
        self.activity='sitting'
        return (self.color,self.activity)
    
# Write your code here
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8
import math
class Cylinder:
    radius=5.0
    height=18.0
    def __init__(self,radius,height):
        print("Default radius={} and height={}".format(Cylinder.radius,Cylinder.height))
        self.radius=float(radius)
        self.height=float(height)
        print('Updated:radius={} and height={}'.format(self.radius,self.height))
        Cylinder.radius=self.radius
        Cylinder.height=self.height
    @classmethod
    def swap(cls,hgt,rad):
        radius=rad
        height=hgt
        return(cls(radius,height))
    @classmethod
    def changeFormat(cls,string):
        lst=string.split('-')
        radius,height=lst[0],int(lst[1])
        return cls(radius,height)
    @staticmethod
    def area(radius,height):
        print('Area:',radius*2*math.pi*(radius+height))
    @staticmethod
    def volume(radius,height):
        print('Volume:',radius*math.pi*(radius)*height)
# Write your code here
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10
class SultansDine:
    num_of_branch=0
    total_sell=0
    lst=[]
    def __init__(self,name,branch_sell=0):
        self.name=name
        self.branch_sell=branch_sell
        SultansDine.num_of_branch+=1
    def sellQuantity(self,quant):
        if quant<10:
           self.branch_sell=quant*300
        elif 11<quant<20:
            self.branch_sell=quant*350
        else:
            self.branch_sell = quant * 400
        x = (self.name, self.branch_sell)
        SultansDine.lst.append(x)
        SultansDine.total_sell+=self.branch_sell
        return self.branch_sell
    def branchInformation(self):
        print('Name:',self.name)
        print('Branch sell:',self.branch_sell,'Taka')
    @classmethod
    def details(cls):
        print('Total number of branch(s):',SultansDine.num_of_branch)
        print('Total sell:',SultansDine.total_sell,'Taka')
        for i in SultansDine.lst:
            print('Branch Name:',i[0],', Branch sell:',i[1])
            if SultansDine.total_sell!=0:
                y=(i[1]/SultansDine.total_sell)*100
                percentage="{:.2f}".format(y)
                print("Branch consists of total sell's:",percentage+'%')

# Write your code here
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
