```python
#Task 1
class Calculator:
    def __init__(self,value1,operator,value2):
        self.value1=value1
        self.value2=value2
        self.operator=operator
        print("Let's calculate!")
        print('Value 1:',self.value1)
        print('Operator:',operator)
        print('Value 2:',self.value2)
        
    def add(self):
        self.result=self.value1+self.value2
        print('Result:',self.result)
    def subtract(self):
        self.result=self.value1-self.value2
        print('Result:',self.result)
    def multiply(self):
        self.result=self.value1*self.value2
        print('Result:',self.result)
    def divide(self):
        self.result=self.value1/self.value2
        print('Result:',self.result)
        
c1 =Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#task 4
class Vehicle:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def moveUp(self,y=0):
        self.y=y+1
        return self.y
    def moveDown(self,y=0):
        self.y=self.y-1
        return self.y
    def moveLeft(self,x=0):
        self.x=x-1
        return self.x
    def moveRight(self,x=0):
        self.x=self.x+1
    def print_position(self):
        print('(',self.x,',',self.y,')')
          
    
# Write your class here
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#Task5
class Shape:
  def __init__(self,name,d1,d2):
    self.name=name
    self.d1=d1
    self.d2=d2

  def area (self):
    if self.name=="Triangle" or self.name=="Rhombus":
      area= 0.5*self.d1*self.d2
    elif self.name== "Square" or self.name== "Rectangle":
      area= self.d1*self.d2
    else:
      area="shape unknown"
    print("Area: ",area)

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```
